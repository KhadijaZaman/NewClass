# **Outlier Detection**

```python
# Step 6: Outlier Detection

# Define a function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Check for outliers in each numerical feature
numerical_features = ['Age', 'Usage', 'Fitness', 'Income', 'Miles']

for feature in numerical_features:
    outliers = detect_outliers_iqr(data, feature)
    print(f"\nOutliers detected in {feature}:")
    print(outliers)
    print(f"Number of outliers in {feature}: {len(outliers)}")
```

# **Conditional Probabilities**

```python
# Step 7: Conditional Probabilities

# 1. Percentage of customers purchasing each product
product_counts = data['Product'].value_counts(normalize=True) * 100
print("Percentage of customers purchasing each product:")
print(product_counts)
```

```python
# 2. Product vs Gender
print("\nPercentage of male and female customers purchasing each product:")
gender_product = pd.crosstab(data['Gender'], data['Product'], normalize='columns') * 100
print(gender_product)
```

```python
# 3. Product vs Age
print("\nPercentage of customers in their 20s and 30s purchasing each product:")
data['Age_Group'] = pd.cut(data['Age'], bins=[0, 20, 30, 40, 50], labels=['Teens', '20s', '30s', '40s'])
age_product = pd.crosstab(data['Age_Group'], data['Product'], normalize='columns') * 100
print(age_product)
```

```python
# 4. Product vs Income
low_income_threshold = 40000
high_income_threshold = 70000

low_income = data[data['Income'] <= low_income_threshold]
high_income = data[data['Income'] >= high_income_threshold]

print("\nPercentage of low-income customers purchasing each product:")
print(low_income['Product'].value_counts(normalize=True) * 100)

print("\nPercentage of high-income customers purchasing KP781:")
print((high_income['Product'] == 'KP781').mean() * 100)
```

```python
# 5. Product vs Fitness
print("\nPercentage of customers with Fitness Level 5 purchasing KP781:")
fitness_level_5 = data[(data['Fitness'] == 5)]
print((fitness_level_5['Product'] == 'KP781').mean() * 100)
```

```python
# 6. Product vs Marital Status
print("\nPercentage of partnered customers using treadmills:")
partnered_customers = data[data['MaritalStatus'] == 'Partnered']
print(partnered_customers['Product'].value_counts(normalize=True) * 100)
```
